```python
import pandas as pd
import numpy as np
```

```python
basics_url = "https://datasets.imdbws.com/title.basics.tsv.gz"
ratings_url = "https://datasets.imdbws.com/title.ratings.tsv.gz"
akas_url = "https://datasets.imdbws.com/title.akas.tsv.gz"
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
ratings = pd.read_csv(ratings_url, sep='\t', low_memory=False)
```

```python
akas = pd.read_csv(akas_url, sep='\t', low_memory=False)
```

# Basics Preprocessing

```python
basics
```

```python
basics.replace({'\\N':np.nan}, inplace = True)
```

```python
basics.dropna(subset=['runtimeMinutes'], inplace = True)
```

```python
basics
```

```python
basics.dropna(subset=['genres'], inplace = True)
```

```python
basics
```

```python
basics_movie = basics[~(basics['titleType'] == 'movie')].index
basics.drop(basics_movie, inplace = True)
```

```python
basics
```

```python
basics['startYear'].dtypes
```

```python
basics['startYear'].value_counts().sort_values().head(20)
```

```python
basics['startYear'] = basics['startYear'].astype(float)
```

```python
basics_years = basics[(basics['startYear'] < 2000) | (basics['startYear'] > 2022)].index
basics.drop(basics_years, inplace = True)
```

```python
basics
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
basics
```

# Akas

```python
akas.replace({'\\N':np.nan}, inplace = True)
```

```python
akas
```

```python
akas.dropna(subset=['region'], inplace = True)
```

```python
# Exclude movies that are included in the documentary category.
akas_us = akas['region'].str.contains('US',case=False)
akas = akas[akas_us]
```

```python
akas
```

```python
akas.info()
```

# Basics in US

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers =basics['tconst'].isin(akas['titleId'])
keepers
```

```python
basics = basics[keepers]
basics
```

```python
basics.info
```

# Ratings

```python
ratings.replace({'\\N':np.nan}, inplace = True)
```

```python
ratings
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
keepers_ratings =ratings['tconst'].isin(akas['titleId'])
keepers_ratings
```

```python
ratings = ratings[keepers_ratings]
ratings
```

```python
ratings.info()
```
